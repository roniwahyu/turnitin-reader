```python
from PyPDF2 import PdfReader
import pdfplumber
from pdf2image import convert_from_path
import pytesseract
from PIL import Image
import os

# Path to the PDF file
pdf_path = "dataset/similarity+index+8244.pdf"

# Directory to save images
output_dir = "hasil/pdf_images"
os.makedirs(output_dir, exist_ok=True)

# Set the Tesseract executable path (adjust this according to your Tesseract installation path)
# Uncomment and set the correct path if needed
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Windows
# pytesseract.pytesseract.tesseract_cmd = '/usr/local/bin/tesseract'  # macOS/Linux (example)

# Function to extract text from PDF using PdfReader
def extract_text_pypdf2(pdf_path):
    pdf = PdfReader(open(pdf_path, "rb"))
    for page_num, page in enumerate(pdf.pages):
        text = page.extract_text()
        print(f"Page {page_num + 1} Text:\n{text}\n")

# Function to extract text and images from PDF using pdfplumber
def extract_with_pdfplumber(pdf_path, output_dir):
    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            text = page.extract_text()
            print(f"Page {page_num + 1} Text:\n{text}\n")
            
            if page.images:
                for img_num, img in enumerate(page.images):
                    # Get the bounding box of the image
                    bbox = (img['x0'], img['top'], img['x1'], img['bottom'])
                    # Crop the image from the page
                    cropped_image = page.within_bbox(bbox).to_image()
                    # Save the image
                    image_path = os.path.join(output_dir, f"page_{page_num + 1}_img_{img_num + 1}.png")
                    cropped_image.save(image_path,dpi=300)
                    print(f"Saved image: {image_path}")
                    # Perform OCR on the image
                    img_text = pytesseract.image_to_string(Image.open(image_path))
                    print(f"Text from image {img_num + 1} on page {page_num + 1}:\n{img_text}\n")

# Function to convert PDF pages to images and save as PNGs at 300 DPI
def convert_pages_to_images(pdf_path, output_dir, dpi=300):
    images = convert_from_path(pdf_path, dpi=dpi)
    for i, image in enumerate(images):
        image_file = os.path.join(output_dir, f"page_{i + 1}.png")
        image.save(image_file, "PNG", dpi=(dpi, dpi))
        print(f"Saved {image_file} at {dpi} DPI")
        # Perform OCR on the saved image
        text = pytesseract.image_to_string(Image.open(image_file))
        print(f"Text from {image_file}:\n{text}\n")

# Run all extraction functions
extract_text_pypdf2(pdf_path)
extract_with_pdfplumber(pdf_path, output_dir)
convert_pages_to_images(pdf_path, output_dir, dpi=300)
```
